```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

paises= ['estados unidos','españa','mexico','rusia','japon']
ventas= [25,32,34,20,25]

fig, ax = plt.subplots()
ax.set_ylabel('ventas')
ax.set_title('cantidad de ventas por pais')
plt.bar(paises,ventas)
plt.savefig('barras_simple.png')
plt.show
```

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 50+ np.random.randn(22)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 197, where=(ys > 197), facecolor='g', alpha=0.5)

plt.title("Visualización")
plt.show()
  
```

```python
import matplotlib.pyplot as plt

labels = 'S1', 'S2', 'S3'
sections = [100, 90, 80]
colors = ['c', 'g', 'y']

plt.pie(sections, labels=labels, colors=colors,
        startangle=10,
        explode = (0, 0.2, 0),
        autopct = '%1.2f%%')

plt.axis('equal') # Try commenting this out.
plt.title('Gráfico de PIE')
plt.show()
```
